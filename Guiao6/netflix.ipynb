Quantos realizadores de programas de televisão diferentes contêm os dados? Quais os 
realizadores de TV que também têm trabalhos em cinema? Encontre todas as linhas do 
realizador "Robert Rodriguez" e tipo "Movie". 

```python
directors_n = nf[nf.tipo=='TV Show'].realizador.unique().__len__()-1
print(directors_n)
directors_tv_movie_n = nf[nf[nf.tipo=='Movie']  == nf[nf.tipo=='TV Show'].sort_index(inplace=True)]
nf2 =  nf[nf.realizador.notna() ]
directors_tv_movie=[]
for director in nf2.realizador.to_list():
    nf3_d = nf[nf.realizador==director]
    types = nf3_d.tipo.drop_duplicates().to_list()
    if len(types)>1:
        directors_tv_movie.append(director)

print(directors_tv_movie) # lista de diretores com tv show e movies
nf_temp = nf.set_index('realizador')
all_lines = nf_temp.loc[directors_tv_movie,:] # df com esses diretores e suas respectivas linhas da df original
print(all_lines)


rebert_rodriguez_movies = nf[(nf.realizador=='Robert Rodriguez') & (nf.tipo=='Movie')]
rebert_rodriguez_movies
```
