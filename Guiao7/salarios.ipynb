# Salarios

Crie os seguintes gráficos utilizando os dados no ficheiro Salaries.csv, que contém informação sobre 
salários de Professores Universitários dos EUA: 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
filename = 'Salaries.csv'
plt.style.use('ggplot')
```

1. Histogramas para o salário com Matplotlib e com Seaborn. 

```python
salaries = pd.read_csv(filename)
salaries = salaries.astype({'rank':'category','discipline':'category','phd':'int8','sex':'category','salary':'int32'})
fig, ax1 = plt.subplots(1)
fig2, ax2 = plt.subplots(1)
ax1.hist(data=salaries,x='salary',color='g',bins=8)
fig2 = sns.histplot(data=salaries,x='salary',ax=ax2,color='k',bins=8)
salaries
```

2. Gráfico de barras verticais com o número de professores em cada rank. 


```python
fig = sns.barplot(data=salaries,x='rank',y='salary')
```

3. Gráfico de barras horizontais com o número de professores em cada rank separados por género 
(sex). 

```python
sns.countplot(data=salaries,y='rank',hue='sex')
```

4. Boxplots com salário em função de rank e sex.

```python
sns.boxplot(data=salaries,x='rank',y='salary',hue='sex')
```
