# Salarios

Crie os seguintes gráficos utilizando os dados no ficheiro Salaries.csv, que contém informação sobre 
salários de Professores Universitários dos EUA: 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
filename = 'Salaries.csv'
plt.style.use('ggplot')
```

1. Histogramas para o salário com Matplotlib e com Seaborn. 

```python
salaries = pd.read_csv(filename,usecols=['rank','discipline','service','sex','salary'],)
salaries = salaries.astype({'rank':'category','discipline':'category','sex':'category','salary':'int32'},)
salaries.columns = ['rank','disciplina','serviço','sexo','salario']
fig, axs = plt.subplots(1)
fig1, axs1 = plt.subplots(1)
axs.hist(data=salaries,x='salario',color='y',bins=8)
axs.set_xlabel('Salario')
axs.set_title('Histograma salario')
axs.set_ylabel('Numero de professores(as)')
fig1 = sns.histplot(data=salaries,x='salario',color='k',ax=axs1,bins=8)
fig1.set(ylabel='Numero de professores(as)',title='Histograma salario')
```

2. Gráfico de barras verticais com o número de professores em cada rank. 


```python
fig2 = sns.barplot(data=salaries,x='rank',y='salario')
fig2.set(title='Numero de professores(as) por cada rank')
```

3. Gráfico de barras horizontais com o número de professores em cada rank separados por género 
(sex). 

```python
fig3 = sns.countplot(data=salaries,y='rank',hue='sexo')
fig3.set(xlabel='Numero de professores(as)',title='Número de professores(as) em cada rank separados por género' )
```

4. Boxplots com salário em função de rank e sex.

```python
fig4 = sns.boxplot(data=salaries,x='rank',y='salario',hue='sexo')
fig4.set(title='Salário em função de rank e sex')
```
