# Análise de Resultados num Exame 

Usando os dados ficheiro ETP.xls, usando Seaborn responda graficamente às seguintes questões 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
filename = 'ETP.xls'
plt.style.use('ggplot')
```

```python
etp = pd.read_excel(io=filename,sheet_name='FINAL')
etp = etp[etp['TP'].isin(['TP1','TP2'])]
etp = etp.drop(columns=['Unnamed: 3'])
etp.head(3)
```

1. Como se distribuem as classificações? 

```python
fig , ax1 = plt.subplots(1)
sns.histplot(data=etp,x='CLASSIF',bins=9,ax=ax1)
#fig.set_xticklabels(fig.get_xticks(),rotation=60)
```

2. Compare as médias das várias turmas criando o gráfico que considere mais adequado. 


```python
fig1,ax1 = plt.subplots()
fig2,ax2 = plt.subplots()

fig1 = sns.boxplot(data=etp,y='CLASSIF',x='P' , ax=ax1)
fig2 = sns.boxplot(data=etp,y='CLASSIF',x='TP' ,ax=ax2)
```

Aprofunde a comparação dos resultados das várias turmas práticas, incluindo como fator 
adicional a turma TP do aluno.

```python
fig3 = sns.barplot(etp,y='CLASSIF',x='P', hue='TP')
fig3.set(xlabel='Turma P', ylabel='Classificação')
```

4. Compare em mais detalhes as turmas práticas e TP usando boxplots. 


```python
fig4,ax4 = plt.subplots(1)
fig5,ax5 = plt.subplots(1)
fig4 = sns.boxplot(etp,y='CLASSIF',x='P',ax=ax4)

fig5 = sns.boxplot(etp,y='CLASSIF',x='TP')
```

Compare o desempenho dos vários cursos tendo em conta a turma TP dos alunos.

```python
sns.barplot(etp,y='CLASSIF',x='CURSO',hue='TP')
```
