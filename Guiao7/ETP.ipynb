# Análise de Resultados num Exame 

Usando os dados ficheiro ETP.xls, usando Seaborn responda graficamente às seguintes questões 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
filename = 'ETP.xls'
plt.style.use('ggplot')
```

```python
etp = pd.read_excel(io=filename,sheet_name='FINAL')
etp = etp[etp['TP'].isin(['TP1','TP2'])]
etp = etp.drop(columns=['Unnamed: 3'])
etp.head(3)
```

1. Como se distribuem as classificações? 

```python

fig = sns.histplot(data=etp,x='CLASSIF',bins=9,)
fig.set(title='Histograma classificações',xlabel ='Classificações',ylabel='Numero de alunos')
fig.get_figure().savefig('../Guiao7/images/ETP/ex1.jpg')
```

2. Compare as médias das várias turmas criando o gráfico que considere mais adequado. 


```python
fig2_p,ax2=plt.subplots(1)
fig2_tp,ax22=plt.subplots(1)


fig2_p = sns.barplot(data=etp,y='CLASSIF',x='P',ax=ax2)
fig2_p.set(title='Medias das classificações turmas das praticas',xlabel='Turma pratica',ylabel='Classificação')
fig2_p.get_figure().savefig('../Guiao7/images/ETP/ex2_p.jpg')



fig2_tp = sns.barplot(data=etp,y='CLASSIF',x='TP', ax=ax22)
fig2_tp.set(title='Medias das classificações turmas das teorica-praticas',xlabel='Turma teorica-pratica',ylabel='Classificação')
fig2_tp.get_figure().savefig('../Guiao7/images/ETP/ex2_tp.jpg')



```

Aprofunde a comparação dos resultados das várias turmas práticas, incluindo como fator 
adicional a turma TP do aluno.

```python
fig3,ax3 = plt.subplots(1)
fig3 = sns.barplot(etp,y='CLASSIF',x='P', hue='TP',ax=ax3)
fig3.set(xlabel='Turma pratica', ylabel='Classificação',
         title='Resultados das turmas praticas por turma teorica-pratica')
ax3.legend(title='Turmas teorica-praticas')
fig3.get_figure().savefig('../Guiao7/images/ETP/ex3.jpg')
```

4. Compare em mais detalhes as turmas práticas e TP usando boxplots. 


```python
fig4_p,ax4_p = plt.subplots(1)
fig4_tp,ax4_tp = plt.subplots(1)


fig4_p = sns.boxplot(etp,y='CLASSIF',x='P',ax=ax4_p)
fig4_p.set(title='Medias das classificações turmas das praticas',xlabel='Turma pratica',ylabel='Classificação')
fig4_p.get_figure().savefig('../Guiao7/images/ETP/ex4_p.pdf')






fig4_tp = sns.boxplot(etp,y='CLASSIF',x='TP')
fig4_tp.set(title='Medias das classificações turmas das teorica-praticas',xlabel='Turma teorica-pratica',ylabel='Classificação')
fig4_tp.get_figure().savefig('../Guiao7/images/ETP/ex4_tp.pdf')
```

Compare o desempenho dos vários cursos tendo em conta a turma TP dos alunos.

```python
fig5, ax5 = plt.subplots(1)

fig5 = sns.barplot(etp,y='CLASSIF',x='CURSO',hue='TP',ax=ax5)
fig5.set(title='Media das classificações por curso e por turma TP',ylabel='Classificação',xlabel='Id do curso')
fig5.get_figure().savefig('../Guiao7/images/ETP/ex5.svg')
```
